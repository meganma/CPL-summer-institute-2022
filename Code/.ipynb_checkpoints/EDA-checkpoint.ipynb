### Import Housing Inventory Count Data (Source: https://www.hudexchange.info/resource/3031/pit-and-hic-data-since-2007/)

```python
os.chdir('../SourceData/cleanedHIC')
```

```python
hic = pd.read_csv('hic_cleaned.csv',dtype={'Address1': str, 
                                           'Address2':str,
                                           'Geo Code':str,
                                           'City':str, 
                                           'Notes':str, 
                                           'housingType':str,
                                           'HMIS Org ID':str,
                                           'HMIS Project ID':str,
                                           'HIC Date':str,
                                           'County':str,
                                            'Total Year Round Beds': float} ) 
hic.head()
```

Add shelter occupancy rates (PIT Count/Total Beds)

```python
occupancy = []
for i in hic.index:
    pit = hic['PIT Count'][i]
    beds = hic['Total Year Round Beds'][i]
    if(pit == 0 or beds == 0):
        occupancy.append(np.nan)
    else:
        occupancy.append(pit/beds)
hic['Occupancy Rates'] = occupancy
```

```python
hic['Occupancy Rates']
```

```python
HIC_PIT_count_2012 = hic[hic['Year']==2012].groupby('State')[['PIT Count']].sum()
HIC_PIT_count_2012.reset_index(inplace=True)
HIC_PIT_count_2012.rename(columns={'PIT Count': "HIC PIT count 2012"},inplace = True)
HIC_PIT_count_2012.head()
```

```python
HIC_PIT_count_2020 = hic[hic['Year']==2020].groupby('State')[['PIT Count']].sum()
HIC_PIT_count_2020.reset_index(inplace=True)
HIC_PIT_count_2020.rename(columns={'PIT Count': "HIC PIT count 2020"},inplace = True)
HIC_PIT_count_2020.head()
```

```python
HIC_PIT_count_2021 = hic[hic['Year']==2021].groupby('State')[['PIT Count']].sum()
HIC_PIT_count_2021.reset_index(inplace=True)
HIC_PIT_count_2021.rename(columns={'PIT Count': "HIC PIT count 2021"},inplace = True)
HIC_PIT_count_2021.head()
```

## Single Variable Regression Plotter

```python
def regressdf(df,x_col,y_col,point_labels,x_label,y_label,title):
    '''
    Args: 
    df: DataFrame with columns you want to plot using linear regression
    x_col: independent var column name
    y_col: dependent var column name
    point_labels: data point label column name (ex: 'State')
    
    x_label: what to label the plot x axis 
    y_label: what to label the plot y axis
    title: what to label the plot title 
    '''
    temp_df = df[[point_labels,x_col,y_col]].dropna()
    temp_df.reset_index(inplace = True, drop = True)

    X = temp_df[x_col]
    Y = temp_df[y_col]
    n = temp_df[point_labels]

    fig, ax = plt.subplots(dpi = 300,figsize=(6, 3))
    ax.scatter(X, Y)

    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    ax.set_title(title)
    for i, txt in enumerate(n):
        ax.annotate(txt, (X[i], Y[i]))

    #linear regression
    regression = stats.linregress(X, Y)
    print('''Slope: {}
    Intercept: {}
    Std Error: {}
    R^2: {}'''.format(regression.slope,regression.intercept, regression.stderr, regression.rvalue**2))

    plt.plot(X, regression.intercept + regression.slope*X, label='fitted line')
    #add text on vis
    ax.text(0.15, 0.6, '''Slope: {}
    R-squared: {}
    Std. Error: {}'''.format(round(regression.slope,10),round(regression.rvalue**2,3),round(regression.stderr,10)), horizontalalignment ='center', verticalalignment='center', transform=ax.transAxes)

    plt.show()
```
