```python
import pandas as pd
pd.set_option('display.float_format', lambda x: '%.3f' % x)
import numpy as np
import os
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams['figure.facecolor'] = 'e8e4e4ff'
plt.rcParams['axes.facecolor'] = 'e8e4e4ff'
import statsmodels.api as sm
import seaborn as sns
np.set_printoptions(suppress=True)
pd.set_option('display.float_format', lambda x: '%.5f' % x)
```

```python
os.chdir('../SourceData/regression_variables')
```

```python
regression_df = pd.read_csv('regression_df.csv').dropna()
regression_df.head()
```

```python
regression_df.rename(columns = {'Pecentage of Senate members who are Democrats':'Percentage of Democrats in Senate',
                               'Percentage of House Representatives who are Democrats':'Percentage of Democrats in House',
                               'Percentage of People with incomes 50% or less of poverty threshold':'Percentage of people under 50% of pov. rate'},inplace = True)
```

```python
regression_df['Rent to Income Ratio'] = regression_df['Median Rent (2020)']/regression_df['Median Household Income']
```

```python
regression_df['Ln Homelessness Rate (%)'] = np.log10(regression_df['Homelessness Rate (%)'])
regression_df[['Ln Homelessness Rate (%)']].head()
```

```python
normalized_regression_df=(regression_df-regression_df.mean())/regression_df.std()
normalized_regression_df.head()
```

```python
normalized_regression_df.columns
```

Create column for log of proportion homeless

```python
regression_df['ln Proportion State Overall Homeless,2020'] = np.log10(regression_df['Proportion State Overall Homeless,2020'])
regression_df[['ln Proportion State Overall Homeless,2020']].head()
```

```python
regression_df.columns
```

Reorder Regression Dataframe

```python
regression_df = regression_df[['State', 
                               'State Full Name', 
                               
                               'Bottom-Tier ZHVI', 
                               'Mid-Tier ZHVI', 
                               'Top-Tier ZHVI',
                               'Condo/Coop ZHVI', 
                               'All Single Family ZHVI',
                               'Percentage of Renters',
                               'Median Rent (2020)',
                               'Median Household Income',
                               'Rent to Income Ratio',
                               'Gini Coefficient of Income Inequality',
                               'Percentage of people under 50% of pov. rate',
                               'Percentage of College Grads',
                               'Percentage of people Unemployed', 
                               'Percentage of Households that are Crowded',
                               'Percentage of Households with Children',
                               'Population Density', 
                               'Percentage of Democrats in Senate',
                               'Percentage of Democrats in House', 
                               'Democratic Governer',
                               'Ln Homelessness Rate (%)',
                               'Homelessness Rate (%)' ]]
```

```python
corrMatrix = regression_df.drop(columns = ['All Single Family ZHVI','Ln Homelessness Rate (%)'
                                           ]).corr()
sns.set(rc={"figure.dpi":400, 'savefig.dpi':300,'figure.figsize':(15,8),'axes.facecolor':'e8e4e4ff', 'figure.facecolor':'e8e4e4ff'})

colormap = sns.color_palette("ch:s=.25,rot=-.25", as_cmap=True)
sns.heatmap(corrMatrix, annot=True, mask = np.triu(corrMatrix), cmap=colormap)
plt.show()
```

## Only Condo Coop

```python
regression_df['Condo/Coop ZHVI (in hundred thousands)'] = regression_df['Condo/Coop ZHVI']/100000
```

```python
x = regression_df[['Condo/Coop ZHVI (in hundred thousands)']]
y = regression_df['Ln Homelessness Rate (%)']

x = sm.add_constant(x)

model = sm.OLS(y, x).fit()
predictions = model.predict(x) 


print_model = model.summary()
print(print_model)
```
